```python
def post_engagement_analyzer(likes):
    total_likes = 0
    for like in likes:
        total_likes += like
    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"
    print("Total Likes:", total_likes)
    print("Post Status:", status)

post_engagement_analyzer([200, 300, 250, 300])
```

```python
def medicine_stock_alert(stock):
    status = "Low Stock Alert" if stock < 10 else "Stock Sufficient"
    print("Medicine Stock:", stock)
    print("Status:", status)

medicine_stock_alert(6)
```

```python
def rainfall_checker(rainfall, required):
    total = 0
    for r in rainfall:
        total += r
    avg = total / len(rainfall)
    status = "Adequate Rainfall" if avg >= required else "Inadequate Rainfall"
    print("Average Rainfall:", avg)
    print("Rainfall Status:", status)

rainfall_checker([70, 75, 80, 63], 70)
```

```python
def duplicate_account_detection(usernames):
    unique_users = set(usernames)
    status = "Yes" if len(unique_users) != len(usernames) else "No"
    print("Duplicate Accounts Found:", status)

duplicate_account_detection(["user1", "user2", "user1"])
```

```python
def appointment_eligibility(age):
    status = "Eligible" if age >= 18 else "Not Eligible"
    print("Patient Age:", age)
    print("Eligibility Status:", status)

appointment_eligibility(21)
```

```python
def premium_crop_filter(prices):
    premium = []
    for price in prices:
        if price > 2000:
            premium.append(price)
    print("Premium Crops:", premium)

premium_crop_filter([1500, 2500, 3200, 1800])
```

```python
def application_health_checker(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    print("Error Count:", errors)
    print("System Status:", status)

application_health_checker(7)
```

```python
def transaction_limit_checker(amount):
    status = "Approved" if amount <= 50000 else "Rejected"
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

transaction_limit_checker(60000)
```

```python
def attendance_eligibility(attendance_list):
    total = 0
    for a in attendance_list:
        total += a
    percentage = (total / len(attendance_list))
    status = "Eligible" if percentage >= 75 else "Not Eligible"
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

attendance_eligibility([80, 75, 85, 80])
```

```python
def electricity_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage

bill_amount, usage_status = electricity_bill(250)
print("Total Bill:", bill_amount)
print("Usage Status:", usage_status)
```
